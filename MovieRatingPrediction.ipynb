```python
# Classify Movies Based on Ratings [Excellent, Good, Average]
data.columns
```

```python
def rating(rating):
  if rating >= 7.5:
    return "Excellent"
  elif rating >= 6.0:
    return "Good"
  else:
    return "Average"
```

```python
# This will make a new column for Rating
data["rating_cat"]=data["Rating"].apply(rating)
```
